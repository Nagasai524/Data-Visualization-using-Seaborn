# <center> Scatter plot </center>

#### Scatter plots can also be used for categorical data but they are mainly used for reprsenting numerical data. This scatter plot is plotted using two different function in seaborn. They are 
1. relplot (Figure level function)
2. scatterplot (Axes level function)

#### When the argument "style" is used, different markers are used for representing each unique type of the column that is passed.
#### The "size" parameter is used to define the marker size and it will be based on the column that is passed to the argument.
#### The "style" parameter can be used to display data in an extra dimensions (features). 

```python
plt.style.use('fivethirtyeight')
sns.relplot(x='city_mpg',y='price',data=auto,hue='engine_type',style='fuel_type',size='highway_mpg')
```

#### The above plot is displaying the information of 5 different features of the dataframe. They are
1. city_mpg on x-axis
2. price on y-axis
3. engine_type being displayed using hue parameter with the help of different colors for each engine type.
4. highway_mpg is represented using size parameter
5. fuel_type is denoted using different markers.

```python
# we can specify the custom sizes of the markers using the "sizes" parameter..
sns.relplot(x='city_mpg',y='price',data=auto,hue='engine_type',style='fuel_type',size='highway_mpg',sizes=(20,200))
```

#### All the above plots can also be generated using scatterplot function as follows

```python
sns.scatterplot(x='city_mpg',y='price',data=auto,hue='engine_type',size='highway_mpg',sizes=(20,200))
```

#### In a pairplot the plots between all combinations of features that are passed as inputs will be created. In the diagnol we will end up plotting between same feature and hence we will have histogram of the features in the diagnol. Hence univariate analysis can be found on the diagnol elements and the non-diagnol plots represent Bi-variate Analysis.

#### The KDE or PDF of the variable that is on the both side of the rectangular bar is symetric in nature. In other words, one is the duplicate of the other. This is like displaying the redundant information to the user. This becomes more concern when a hue paramter is used. When hue is used, we will have vilon plots for each unique value of that featue because of which we may run out of space. In such a case, we can enable the splot argument. This argument displays the PDF of two unique values of the feature in the same vilon instead of creating an other.

```python
sns.catplot(x='fuel_type',y='price',hue='engine_location',data=auto,kind='violin')
```

```python
sns.catplot(x='fuel_type',y='price',hue='engine_location',data=auto,kind='violin',split=True)
```

#### The split function will work only when the hue has two different levels. In other words the total number of unique values in the X feature should be equal to two.

# <center> Categorical Categorical Plots </center>
## <center> Heat Map </center>
#### For creating a Heat map, we require pivot table similar to the expected heat map. Hence, Pivot table must be created fast and the created pivot table will be passed as an input argument to the heatmap function
1. cbar=True (enables the presence of color indicator on the right).
2. linewidths = integer (Can be tought of padding to each cell of table.)
3. cmap = theme of colors to be used in the heat map.
4. annot= True (This will insert the value in every box of the heat map. Null values are not replaced with anything.)

```python
x=auto.pivot_table(index='engine_type',columns='number_of_doors',values='price',aggfunc=np.mean)
sns.heatmap(x,annot=True,linewidths=5,cmap='summer')
```

```python
x=auto.pivot_table(index='make',columns='body_style',values='price',aggfunc=np.mean)
sns.heatmap(x,cmap='winter',annot=True,linewidths=0.5,cbar=False)
```

### Heat map for Total number of passenger in a given year and month using Flights dataset of Searborn

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,8))
pivot=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
sns.heatmap(pivot,linewidths=0.5,cmap='viridis',annot=True)
```

# <center> Cluster Maps (Heat Maps + Clustering of Rows and Columns)</center>
### Similar rows and columns are grouped together and the order of grouping is explained by the graphs on the top. We can enable or disable either row clustering or column clustering. When both are disable the cluster map will be equivalent to Heat map. All arguments of Heat map also work with Cluster map
Arguments
-------------
1. z_score = 0 (Normalizes the values to be between -1 and 1)
2. metric = 'euclidean' (The distance metric using which clustering is performed. This can be changed to any kind of distance like minkowski,manhatten and so on.)
3. row_cluster = False (Disables the clustering of rows)
4. col_cluster = False (Disables the clustering of columns)

```python
sns.clustermap(pivot,z_score=0,metric='minkowski',row_cluster=False,annot=True)
```

```python
sns.clustermap(pivot,metric='correlation',col_cluster=False,annot=True)
```

# <center> Multi Panel Categorical plots </center>
## <center> Factor Plots </center>

#### In factor plots, the plot that is being created will be created for all the unique values present in the feature passed to the col attribute.

```python
sns.factorplot(x="engine_type", y="price", col="number_of_doors", data=auto, kind="swarm")
```

#### Plotting all kinds of possible factor plots

```python
kinds=['point', 'bar', 'box', 'violin', 'strip']
for kind in kinds:
    sns.factorplot(x="engine_type", y="price", col="number_of_doors", data=auto, kind=kind)
```
